# Data Preparation

#### Dummy Variables

```python
fuel_type = pd.get_dummies(Car_Price['fueltype'])
```

```python
fuel_type.head()
```

```python
fuel_type = pd.get_dummies(Car_Price['fueltype'], drop_first = True)
```

```python
fuel_type.head()
```

```python
Car_Price = pd.concat([Car_Price, fuel_type], axis = 1)
```

```python
Car_Price.head()
```

```python
Car_Price.drop(['fueltype'], axis = 1, inplace = True)
```

```python
Car_Price.head()
```

```python
symbol = pd.get_dummies(Car_Price['symboling'],drop_first = True)
symbol.head()
```

```python
Car_Price = pd.concat([Car_Price, symbol], axis = 1)
Car_Price.head()
```

```python
Car_Price.drop(['symboling'], axis = 1, inplace = True)
Car_Price.head()
```

```python
aspire = pd.get_dummies(Car_Price['aspiration'],drop_first = True)
Car_Price = pd.concat([Car_Price, aspire], axis = 1)
Car_Price.drop(['aspiration'], axis = 1, inplace = True)
Car_Price.head()
```

```python
door_number = pd.get_dummies(Car_Price['doornumber'],drop_first = True)
Car_Price = pd.concat([Car_Price, door_number], axis = 1)
Car_Price.drop(['doornumber'], axis = 1, inplace = True)
Car_Price.head()
```

```python
car_body = pd.get_dummies(Car_Price['carbody'],drop_first = True)
Car_Price = pd.concat([Car_Price, car_body], axis = 1)
Car_Price.drop(['carbody'], axis = 1, inplace = True)
Car_Price.head()
```

```python
drive_wheel = pd.get_dummies(Car_Price['drivewheel'],drop_first = True)
Car_Price = pd.concat([Car_Price, drive_wheel], axis = 1)
Car_Price.drop(['drivewheel'], axis = 1, inplace = True)
Car_Price.head()
```

```python
engine_location = pd.get_dummies(Car_Price['enginelocation'],drop_first = True)
Car_Price = pd.concat([Car_Price, engine_location], axis = 1)
Car_Price.drop(['enginelocation'], axis = 1, inplace = True)
Car_Price.head()
```

```python
engine_type = pd.get_dummies(Car_Price['enginetype'],drop_first = True)
Car_Price = pd.concat([Car_Price, engine_type], axis = 1)
Car_Price.drop(['enginetype'], axis = 1, inplace = True)
Car_Price.head()
```

```python
cylinder_number = pd.get_dummies(Car_Price['cylindernumber'],drop_first = True)
Car_Price = pd.concat([Car_Price, cylinder_number], axis = 1)
Car_Price.drop(['cylindernumber'], axis = 1, inplace = True)
Car_Price.head()
```

```python
fuel_system = pd.get_dummies(Car_Price['fuelsystem'],drop_first = True)
Car_Price = pd.concat([Car_Price, fuel_system], axis = 1)
Car_Price.drop(['fuelsystem'], axis = 1, inplace = True)
Car_Price.head()
```

```python
Car_Name = pd.get_dummies(Car_Price['CarName'],drop_first = True)
Car_Price = pd.concat([Car_Price, Car_Name], axis = 1)
Car_Price.drop(['CarName'], axis = 1, inplace = True)
Car_Price.head()
```

#### Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(Car_Price, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
df_train
```

#### Rescaling the features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
num_vars = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

## Building a model

#### Fit a regression line through the training data using statsmodels

```python
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train[['enginesize']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Check the parameters obtained

lr.params
```

```python
# Let's visualise the data with a scatter plot and the fitted regression line
plt.scatter(X_train_lm.iloc[:, 1], y_train)
plt.plot(X_train_lm.iloc[:, 1], 0. -0.072 + 1.210676*X_train_lm.iloc[:, 1], 'r')
plt.show()
```

```python
print(lr.summary())
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 10)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
#columns which rfe did not select
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

```python
#twelve is insignificant in presence of other variables; can be dropped
X_train_new = X_train_rfe.drop(["twelve"], axis = 1)
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_lm = sm.add_constant(X_train_new)
```

```python
lm = sm.OLS(y_train,X_train_lm).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
print(lm.summary())
```

```python
X_train_new.columns
```

```python
X_train_new = X_train_new.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
